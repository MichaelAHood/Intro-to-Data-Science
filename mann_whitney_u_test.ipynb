```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu
import matplotlib.pyplot as plt

```

```python
subway_df = pd.read_csv('turnstile_weather_v2.csv')
```

```python
features = subway_df[['rain', 'precipi', 'hour', 'meantempi']]
```

```python
dummy_units = pd.get_dummies(subway_df['UNIT'], prefix='unit')
features = features.join(dummy_units)
```

```python
raining_df = subway_df[subway_df.rain == 1]
not_raining_df = subway_df[subway_df.rain == 0]
```

```python
ridership_df = pd.DataFrame({'riders_when_raining': raining_df.ENTRIESn_hourly, 
                             'riders_when_not_raining': not_raining_df.ENTRIESn_hourly})
```

```python
# plot the distributions in a histogram to see what the distributions look like

pd.options.display.mpl_style = 'default'
figure = ridership_df.plot(kind='hist', alpha=0.8, bins=100, stacked=True, xlim=(1, 8000), title="Ridership based on precipitation")
figure.set_ylabel('Count')
figure.set_xlabel('Number of Riders per Hour')
```

```python
# run a Mann-Whitney U test to see if there is a difference in distributions
# this outputs a one-tailed test stat

U, p = mannwhitneyu(ridership_df['riders_when_raining'], ridership_df['riders_when_not_raining'], use_continuity=True)
print "U stat = %f and p-value = %f" % (U, p)
print "The mean riders_when_not_raining = %f and the mean riders_when_raining = %f" % (ridership_df['riders_when_raining'].mean(), 
                                                                                       ridership_df['riders_when_not_raining'].mean()) 
print "The median riders_when_not_raining = %f and the median riders_when_raining = %f" % (ridership_df['riders_when_raining'].median(), 
                                                                                       ridership_df['riders_when_not_raining'].median()) 
```

```python
# now let's check to see if there is a difference in ridership when we segment by weekday or weekend

wkday_riders = raining_df[raining_df.weekday == 1]
wkend_riders = raining_df[raining_df.weekday == 0]
```

```python
no_rain = subway_df[(subway_df.rain == 0) & (subway_df.weekday == 1)]
rain = subway_df[(subway_df.rain == 1) & (subway_df.weekday == 1)]
len(no_rain), len(rain)
```

```python
wkday_rain_riders = pd.DataFrame({'rain': rain.ENTRIESn_hourly, 'no_rain': no_rain.ENTRIESn_hourly})
```

```python
figure = wkday_rain_riders.plot(kind='hist', alpha=0.8, bins=100, stacked=True, xlim=(1, 8000), title="Ridership on weekdays")
figure.set_ylabel('Count')
figure.set_xlabel('Number of Riders per Hour')
```

```python
U, p = mannwhitneyu(wkday_rain_riders['rain'], wkday_rain_riders['no_rain'], use_continuity=True)
print "For Weekdays the Mann-Whitney U test results are:"
print "U stat = %f and p-value = %f" % (U, p)
print "The mean no_rain = %f and the mean rain = %f" % (wkday_rain_riders['rain'].mean(), 
                                                        wkday_rain_riders['no_rain'].mean())
print "The median no_rain = %f and the median rain = %f" % (wkday_rain_riders['rain'].median(), 
                                                        wkday_rain_riders['no_rain'].median())
```

```python
wkend_no_rain = subway_df[(subway_df.rain == 0) & (subway_df.weekday == 0)]
wkend_rain = subway_df[(subway_df.rain == 1) & (subway_df.weekday == 0)]
len(wkend_no_rain), len(wkend_rain)
```

```python
wkend_rain_riders = pd.DataFrame({'rain': wkend_rain.ENTRIESn_hourly, 'no_rain': wkend_no_rain.ENTRIESn_hourly})
```

```python
figure = wkend_rain_riders.plot(kind='hist', alpha=0.8, bins=100, stacked=True, xlim=(1, 8000), title="Ridership on weekends")
figure.set_ylabel('Count')
figure.set_xlabel('Number of Riders per Hour')
```

```python
U, p = mannwhitneyu(wkend_rain_riders['rain'], wkend_rain_riders['no_rain'], use_continuity=True)
print "For Weekends the Mann-Whitney U test results are:"
print "U stat = %f and p-value = %f" % (U, p)
print "The mean no_rain = %f and the mean rain = %f" % (wkend_rain_riders['rain'].mean(), 
                                                        wkend_rain_riders['no_rain'].mean())
print "The median no_rain = %f and the median rain = %f" % (wkend_rain_riders['rain'].median(), 
                                                        wkend_rain_riders['no_rain'].median())
```
