```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
# Read the data into a pandas dataframe.

data = pd.read_csv("titanic_data.csv")
```

```python
# It looks like the "Survived" column is a "0" for "did not survive"
# and is a "1" for "did survive".
# so, to double check this, the mean of survived times the total passengers should be 
# the count of all "1's" in the "Survived column

data.Survived.mean() * 891 == data.Survived[data.Survived == 1].count()
```

```python
# Now, let's run a simple hueristic based on the traditional 
# nautical survival procedure called "women and children first".
# Or, to take it a bit farther, assume that men do not survive while women and children do.

# To test this, we will store the predictions in a dict like so, {"passenger_ID" : 0 } if they did not survive
# or, conversely {"passenger_ID" : 1 } if they did survive.  We can then compare the results against the 
# dataset to see how good the prediction model is.

# First, create a datastruct to store the actual survivor data for comparison purposes.

actual = {}
```

```python
# now go through each passenger and assign their {PassengerId, Surived} key/value pair to the dict

for i in range(len(data)):
    actual[data.PassengerId.ix[i]] = data.Survived.ix[i]
```

```python
# Now, I will impliment the simple hueristic, i.e. the one in which I assume that
# men died, and women and children survived.

def simple_heuristic(data):
    
    predictions = {}
    
    for i in range(len(data)):
               
        if data.ix[i].Sex == "male":
            predictions[data.ix[i].PassengerId] = 0
        else:
            predictions[data.ix[i].PassengerId] = 1
            
    return predictions
```

```python
#predictions = simple_heuristic(data)
```

```python
# Now that we have our predictions, we need to create a procedure that will comapre the 
# predicted results against the actual results

def evaluate_predictions(predictions, actual):
    
    num_correct = 0
    
    for passenger in predictions:
        
        if predictions[passenger] == actual[passenger]:
            num_correct += 1
    
    # don't forget to convert to float or the answer will be 0... or 1 if I really nail it!
    result = num_correct / float(len(predictions)) * 100 
    
    return "Your model achieved an accuracy of: %f percent" % result
        
```

```python
# 78.7% accuracy for a first run.

predictions = simple_heuristic(data)

print evaluate_predictions(predictions, actual)
```

```python
def custom_heuristic(data):
    
    predictions = {}
       
    for passenger_index, passenger in data.iterrows():
       
        passenger_id = passenger['PassengerId']
       
        if passenger['Pclass'] == 1 and passenger['Age'] < 18:
            predictions[passenger_id] = 1
       
        if passenger['Sex'] == 'female':
            if passenger['Pclass'] == 3 and passenger['Parch'] > 0:
                predictions[passenger_id] = 0
            else:
                predictions[passenger_id] = 1 
        
        else:
            predictions[passenger_id] = 0
    
    return predictions
```

```python
# This more complex model is able to achieve an accuracy of 80.36%

predictions = custom_heuristic(data)

print evaluate_predictions(predictions, actual)
```
