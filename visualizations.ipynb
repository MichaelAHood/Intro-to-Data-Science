```python
# put each unit into a list

units = []
lat = [] # index = 13, 14
lon = []

for row in subway_df.iterrows():
    if row[1].UNIT not in units:
        units.append(row[1].UNIT)
        lat.append(row[1].latitude)
        lon.append(row[1].longitude)
        
   
```

```python
# put the sum of total riders by unit into a list

total_riders_unit = []
for unit in units:
    sum_riders = subway_df.ENTRIESn_hourly[subway_df.UNIT == unit].sum()
    total_riders_unit.append(sum_riders)
```

```python
# normalize the total riders

mu = unit_df.total_riders.mean()
sigma = unit_df.total_riders.std()

normalized_riders = pd.Series((unit_df.total_riders - mu) / sigma)

len(normalized_riders)
```

```python
unit_df = pd.DataFrame({'total_riders': total_riders_unit, 
                        'normalized_riders': normalized_riders, 
                        'latitude': lat, 
                        'longitude': lon}, index = units)
```

```python
# a plot where the sizing is based on where the value of total_riders falls on a normal distribution
# i.e. the ridership values have been normalized

unit_df.plot(kind='scatter', x='latitude', y='longitude', s=unit_df['normalized_riders']*100)
```
